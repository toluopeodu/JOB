# Track Durations

```python
# The duration of the track is measured in micro seconds, but minutes are more relatable. We therefore create a new column which measures
# the duration of the track in minutes.
converter = 60000

df["Track_Min"] = (df["duration_ms"]  / converter)
```

```python
df['Track_Min']   # This is song duration in minutes
```

```python
df.head(3)
```

```python
track_in_mins =df.groupby('genre')["Track_Min"].mean().sort_values(ascending = False).reset_index()   #We find the mean duration of each song in minutes
track_in_mins
```

```python
# We create a plot of the average duration of songs
sns.set(rc = {'figure.figsize':(10,7)})
sns.barplot(data = track_in_mins, x = 'genre', y='Track_Min', palette= my_palette_w15colors)#"bright")
plt.xticks(rotation=70)
plt.tight_layout()
plt.title("Average Duration of Genre")
plt.xlabel("Genre")
plt.ylabel("Minutes")
```
